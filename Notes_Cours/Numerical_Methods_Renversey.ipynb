### I.2 - Classification of problems in computational photonics

Three types of propblems in computational photonics:

1- Time-domain simulations
> - start with some time-dependent current source **J**(**r**,t)
> - compute the fields **E**(**r**,t) and **H**(**r**,t).

Exemple: Finite-Difference Time-Domaine method (FDTD) and Beam Propagation Method (BPM)

2- Frequency-domain linear response simulations
> - start with and harmonic currect **J**(**r**,t)=**J**(**r**) $e^{i \omega t}$
> - solve the steady-state harmonic problem for **E**(**r**) and **H**(**r**).
> - involve the solving of a linear system of the for Ax=B  to bind x (the filed) where A is a matrix
Exemple: harmonic diffraction Finite Element Method

3- Frequency-domain eigenvalue problem
> - assume source-free configuration
> - search for steady-state harmonic steady-state harmonic problem for E(r) and H(r) with $e^{i \omega t}$ time dependency.
> - involve the solving of a generalized eigenvalue problem of form Ax=Ⲗ²Bx where A and B are matrices (instead of a simple eigenvalue proble Ax=Ⲗ²x). The unknowns are the eigenvalues Ⲗ² and the eigenvectors x.

Remark: in a waveguide problem, the eigen value Ⲗ² is the squared propagation constant β² while in a cavity problem it is a squared frequency ω².

Exemples: Finite Element Method (FEM), Plane Wave Method (PWM)

### Eache of the 3 categories can be split in 2 depending on the geometry of the structures:
1 - periodic structures: --> band diagram for photonic crystal for exemple

2 - finit size structures: --> properties and optimization of photonic devices.


Definition: modal problems, are the frequency-domain eigenvalue problems. The modes being the solutions of the *homogeneous electromagnetic problem*

Justification: Knowning the modes of a waveguide helps to understand diffaction pattern issues and to catch the main properties of photonic structures

### Remarks
>1- All the phenomenons on physics are not linear. In several cases nonlinearities can not be neglected or are even the key effects to be considered.
- The superposition principle is then no longer valid. In these cases, we can keep the above splitting into 3 categories of problems nevertheless the equations to be solved are different and more specific techniques must be used to solve them.
- Most of the times, simplification of Maxwell's equation is needed prior to numerical simulations. Cf "computational photonics and applications".

>2- For frequency-domain eigenvalue problems (modal problems) several numerical methods are available to obtain the eigenvalue equation to be solved (like Ax=Ⲗ²Bx) according to the types of discretization basis they use to reduce the number of unknows/to described the fields.

>3- Sometimes to understand or to catch the physics of one photonic device, only few modes need to be studied, and these modes can be used in *coupled mode theory*

### I.3 - Choice of the discretization basis
#### Finite difference methods
>On a *regular grid*, most of the time a cartesian one, the fields are discretized by their values on the nodes of the grid: $E_n = E(n\Delta x)$ in 1D case.
The derivatives are computed by finite difference on the grid:
- (usually in an explicit way): $\frac{dE_n}{dX}=\frac{dE_{n+1}-dE_{n-1}}{2\Delta x} + \Theta(\Delta x^2)$
- (centered formula for first order derivative)

#### Spectral methods
>The fileds are expressed as a series expansion in a *complete basis set of functions that are soluations of some wave equations*. The series must be truncated to get a finite number of terms to allow the numerical evaluations.

> - The most well known examples is the *plane waves method* (PWM) which uses *Fourier series* in cartesian coordinate system to describe the fields. The terms in the Fourier series are *plane waves that are solutions of the wave equation in homogeneous medium*: $\Delta \mathbf{U}=\frac{1}{v^2}\frac{\partial^2\mathbf{U}}{\partial t^2}$

Using specific approximate trick (supercell approximation), PWM can be used to study finite size structures (with some limitations).
> - Another example is the *multipole method* (MM) which uses *Fourier-Bessel series* to describe the field: the terms in the series invole plane wave modulated by Bessel $J_n$ or Hanker $H_n^{(1)}$ special functions satisfying the *Helmotz equation*.

It is used to study *z*-inveriant structure with and $e^{i \beta z}$ dependency with a finite cross-section in the transverse *(x, y)* or *(x, $\theta$)* plane like microstructured optical fibers (see also boundary-element method paragraph below).

In the transverse plane, using polar coordinates *(x, $\theta$)* in homogeneous regions, $k_⊥$ being the transverse wavenumber such that $k_⊥ = \sqrt{k_0^2 n_{medium}^2 - \beta^2}$.

$V = \sum_{n∈ℤ}{(A_n J_n (k_⊥ r) + B_n H_n^{(1)} (k_⊥ r))} e^{i n \theta}$

Using special boundary conditions, this method can be extended to deal with periodic structures, in this case it is called the *Rayleigh's method*.

- Finite element methods
> The photonic structure is divided into a sed of finite size simple geometric elements (like irregular triangles in 2D or tetrahedra in 3D) called mesh, and the unknown fields are represented on these elements by low-degree polynomials ( not solution of some wave equations).

- Boundary-element methods
> - Not all space is discretized
> - Only the boundaries between homogeneous media are described.
> - The homogeneous region are treated analytically using Helmoltz equation.
> - The discretization can use a finite element or a spectral basis.


Generally speaking:
- Finite difference and spectral methods are easier to implement than FEM or BEM
- FEM are the more versatile methods to studay complex structures including nonlinear ones.

In what follows we will first focus on FEM and espesciallu on its capacity to solve frequency-domain eigenvalue problems.

Secondly, we will describe FDTD. We choose these methods because together they cover a wide field in compuational photonics sutides and they are quite widespread in laboratories and research departments of high-tech photonic companies.

### II.2 - The one-dimensional scalar wave equation

#### Introduction

In this section, we will introduce the first analytical results for the scalar wave equation, then some finit difference numerical schemes.

These results will allow us to describe key concepts in numerical dispersion, numerical phase velocity and critical time-step, and finally numerical stability.
##### II.2.1 - Analytical results for the 1D scalar wave equation

$\frac{\partial^2u}{\partial t^2} = v^2 \frac{\partial^2 u}{\partial x^2}$
with $u = u(x,t)$


It is known that the solutions are the form:


$u(x,t) = F(x+vt) + G(x-vt)$ where $F, G ∈ C^2(R)$

Propagating waves described by F move toward negative x while the ones described by G move toward positive x. 

##### II.2.2 - Dispersion relation

We consider a continuous sinusoidal propagation-wave toward positive solution of the form:


$u(x,t) = e^{i(wt - kw)}$

Subtsituing this expression in Eq. (2), on gets the *dispersion relation*  for the 1D scalqr wave equation:

$k = \pm \frac{\omega}{v}$


We remind that the definition of *phase velocity* is:


$v_p = \frac{\omega}{\mathbf{k}}$


We get in our case:


$v_p = \pm v$

Here $v_p$ does not depend on the frequency, thus the waves are dispersionless.

We now remind the definition of the group velocity v_g:


$v_g = \frac{d\omega}{d\mathbf{k}}$


We get in our case:


$v_g = \pm v$

Again, the velocity is independent of the frequency.



##### II.2.3 - Finite difference approximation of the scalar wave equation

Let us consider a Taylor's series expansion of $u(x, t_n)$ at $ x = x_i + \Delta x$, keeping time fixed:


$u(x_i + \Delta x) = u|_{x_i, t_n} + \Delta x \frac{\partial u}{\partial x}|_{x_i, t_n} + \frac{(\Delta x)^2}{2!} \frac{\partial^2 u}{\partial x^2}|_{x_i, t_n} + \frac{(\Delta x)^3}{3!} \frac{\partial^3 u}{\partial x^3}|_{x_i, t_n} + \frac{(\Delta x)^4}{4!} \frac{\partial^4 u}{\partial x^4}|_{\xi_1, t_n}$

where the last term is the error term and $\xi_1$ $\epsilon$ $[x_i ,x_i + \Delta x]$.

Let us consider a Taylors' series expansion of $u(x, t_n)$ at $x = x_i - \Delta x$, keeping time fixed:


$u(x_i - \Delta x) = u|_{x_i, t_n} - \Delta x \frac{\partial u}{\partial x}|_{x_i, t_n} + \frac{(\Delta x)^2}{2!} \frac{\partial^2 u}{\partial x^2}|_{x_i, t_n} - \frac{(\Delta x)^3}{3!} \frac{\partial^3 u}{\partial x^3}|_{x_i, t_n} + \frac{(\Delta x)^4}{4!} \frac{\partial^4 u}{\partial x^4}|_{\xi_2, t_n}$


where the last term is the error term and $\xi_2$ $\epsilon$ $[x_i ,x_i - \Delta x]$.

Let us consider a Taylors' series expansion of $u(x, t_n)$ at $x = x_i - \Delta x$.


Adding Equations (9) and (10), one gets:


$u(x_i + \Delta x) + u(x_i - \Delta x) = 2 u|_{x_i, t_n} + (\Delta x)^2 \frac{\partial^2 u}{\partial x^2}|_{x_i, t_n} + \frac{(\Delta x)^4}{12} \frac{\partial^4 u}{\partial x^4}|_{\xi_3, t_n}$

This implies:

$\frac{\partial^2 u}{\partial x^2}|_{x_i, t_n} = \frac{(u(x_i + \Delta x) -2 u|_{x_i, t_n} + u(x_i - \Delta x))}{(\Delta x)^2} + \Theta((\Delta x)^2) = \frac{(u|_{x_i+1, t_n} -2 u|_{x_i, t_n} + u|_{x_i-1, t_n})}{(\Delta x)^2} + \Theta((\Delta x)^2)$


where $\Theta((\Delta x)^2)$ is the remainder term.

To follow the usual notations in the field we adopt a supercrippt *n* to specify the time observation point and a subscrit *i* for the space position along *x* axis, eq. (12) is rewritten:


$\frac{\partial^2 u}{\partial x^2}|_{x_i, t_n} = \frac{u^n_{i+1} -2 u^n_{i} +u^n_{i-1}}{(\Delta x)^2} + \Theta((\Delta x)^2)$

We same type of calculus can be done for the second partial derivative, we keep *x* fixed and expand *u* in forward and backward Taylor's expansion in time.

$\frac{\partial^2 u}{\partial t^2}|_{x_i, t_n} = \frac{u^{n+1}_{i} -2 u^n_{i} +u^{n-1}_i}{(\Delta t)^2} + \Theta((\Delta t)^2)$


Subsituing eqs (13-14) in eq. (1), one obtains


$\frac{u^{n+1}_{i} -2 u^n_{i} +u^{n-1}_i}{(\Delta t)^2} + \Theta((\Delta t)^2) = c^2 \frac{u^n_{i+1} -2 u^n_{i} +u^n_{i-1}}{(\Delta x)^2} + \Theta((\Delta x)^2)$



Neglecting the series remainder terms, one have: 


$u^{n+1}_{i} \simeq (c \Delta t)^2 (\frac{u^n_{i+1} -2 u^n_{i} +u^n_{i-1}}{(\Delta x)^2}) +2 u^n_{i} - u^{n-1}_i$

##### Critical time step: the "magic" one

If one choose the special case $c \Delta t = \Delta x$ then, in a **exact way** eq. (16) reduces to:


$u^{n+1}_{i} = u^n_{i+1} +u^n_{i-1} +u^{n-1}_i$


The details of the proof of this properties can be found in Taflove's book, it is based on a simple and direct evaluation of the RHS and LHS terms in Eq. (3) with $u^n_i = F(x_i + ct_n) + G(x_i - ct_n)$.
The "magic" properties come from the cancellation of the time and space remainder terms in eq. (15) coming from eqs. (13-14).

### II.3 - Numerical dispersion relation
#### II.3.1 - Complex wavenumbers

We can follow a similar procedure to the one of section p. 41 to evaluate the *numerical dispersion relation* of the finite-difference scheme given by eq. (16) of the 1D scalar wave solution eq. (1).
We consider a complex exponential wave solution of eq. (1).
We define the wavenumber as possible complex as $\tilde{k} = \tilde{k}_{real} + i \tilde{k}_{imag}$

*this the pure complex number "i" but we use this notation i not to be confused with the x grid subscript *i* use before and after* and plug it into eq. (1):

$u^n_i = e^{i(\omega t_n - \tilde{k} x_i)}$


$u^n_i = e^{i(\omega \Delta t - \tilde{k} i\Delta x)} =  e^{(\tilde{k}_{imag} i\Delta x)}e^{i(\omega \Delta t - \tilde{k}_{real}i\Delta x)}$

Usually, $\tilde{k}$ differs from *k*
The difference is called the *numerical dispersion artifact*.



#### II.3.2 - Three cases 
It can  give rise to numercial wave amplitudes and velocities that are different from the exact ones.

> - If $\tilde{k}_{imag} = 0$ then the wave amplitude is constant.
> - If $\tilde{k}_{imag} < 0$ then the wave amplitude is exponentially decreasing.
> - If $\tilde{k}_{imag} > 0$ then the wave amplitude is exponentially increasing.

We plug expression (18) into eq. (16) to finally obtain after simplification:

$e^{i(\omega \Delta t)} = (\frac{c \Delta t}{\Delta x})^2 (e^{-i(\tilde{k} \Delta x)} -2 + e^{i(\tilde{k} \Delta x)}) +2 -e^{-i(\omega \Delta t)}$

Using Euler's formula, one gets:


$cos(\omega \Delta t) = (\frac{c \Delta t}{\Delta x})^2 (cos(\tilde{k} \Delta x) - 1) + 1$

Inverting the above relation, one obtains the expression for $\tilde{k}$:


$\tilde{k} = \frac{1}{\Delta x} arccos(1 + (\frac{\Delta x}{c \Delta t})^2 (cos(\omega \Delta t)-1))$


Relations (21) and (22) are much more complicated than the simple dispersion relation for the continuous analytical solution of the 1D scalar wave equation (1).


We can use these relations to understand the effects of the numerical discretization on the computed wave properties.


Let us consitder 3 cases:

> - 1) Very fine sampling in time and space
> - 2) Magic time step
> - 3) Dispersive wave propagation

##### 1. Very fine sampling in time and space
Using time step smallness, we got a two-term expansion for the cosinus:


$\tilde{k} = \frac{1}{\Delta x} arccos(1 - \frac{(k \Delta x)^2}{2} )$ where $k = \frac{\omega}{c}$


Then, we use spatial step smallness:


$\tilde{k} = \frac{1}{\Delta x} (k \Delta x) = k$


We see that **we recover the free-space wavenumber** of the full analytical solution **of the initial scalar wave equation**. This implies that both the phase velocity $v_p$ and the group velocity $v_g$ are equal to c regarless of the frequency.


The numerical solution is dispersionless.
**In the limit of vanishing space and time steps, the proposed finite-difference scheme becomes exacts**.

##### 2. Magic time step
We have $c \Delta t = \Delta x$, so we get from eq. (22)
$\tilde{k} = \frac{1}{\Delta x} arccos(1 + (cos(\omega \Delta t)-1))$


$\tilde{k} = \frac{1}{\Delta x} arccos((cos(\omega \Delta t))) = \frac{\omega \Delta t}{c \Delta t} = k$


Again, we recover the free-space wavenumber of the full analytical solution of the scalar wave equation, and the numerical solution is exact even if the spatial and time increments are large.
##### 3. Dispersive wave propagation

We introduce the parameter S known as the *Courant number or numerical stability factor by*:


$S = \frac{c \Delta t}{\Delta x}$

We also define the grid sampling resolution in space cell per free-space wavelengths:


$N_{lambda} = \frac{\lambda_0}{\Delta x}$

We rewrite eq. (22) as:

$\tilde{k} = \frac{N_{lambda}}{\lambda_0} arccos(1 + (\frac{1}{S})^2 + (cos(\frac{2 \pi S}{N_{lambda}})-1))$


$\tilde{k} = \frac{1}{\Delta x} arccos( \xi) =  \frac{1}{\Delta x} (\frac{\pi}{2} - arcsin(\xi)) $

where $\xi = 1 + (\frac{1}{S})^2 + (cos(\frac{2 \pi S}{N_{lambda}})-1)$

It can be shown that the transition between real and complex values of $\tilde{k}$ occurs when $\xi = -1$, the associated value of $N_{lambda}$ denoted by $N^{critical}_{lambda}$ is:


$N^{critical}_{lambda} = \frac{2 \pi S}{cos(1-2S^2)}$

> - 1) For $N_{lambda} > N^{critical}_{lambda}$ (thin grid) $\tilde{k}$ $\epsilon$ $\Re$ so there is no wave attenuation. We can show that $\tilde{v}_p <$ c
> - 2) For $N_{lambda} < N^{critical}_{lambda}$ (coarse grid), $\tilde{k}$ $\epsilon$ C $ - \Re$ so the wave is exponentially decaying. We can show that $\tilde{v}_p$ can be above c.


To clarify this third case (dispersive wave propagationn in p. 57), we now fix S = 0.5, providing $N^{critical}_{lambda} = 3$ and we will investigate the two sub-cases mentioned just above.

### II.4 - Numerical stability
#### II.4.1 - Introduction

We have seen that the choice $\Delta x$ and $\Delta t$ can change the numerical phase velocity within finite-difference grid modeling of the 1D scalar wave equation, and thus the numerical error.


We will show that **another effect** sets a criterium that must be taken into account for $\Delta t$ **to ensure the numerical statibility**.

Let use the following definition:


*An explicit numerical solution of Eq. (1) is* **stable** if it produces a bounded result given a bounded input. The *numerical solution is* **unstable** *if it produces an unbounded result given a bounded input*.


Numerical instability is an undesirable possibility for explicit numerical differential-equation solvers. It causes the computed results to spuriously increase without limit during time-marching.


Here's we will use a quite simple method to study numerical instability. The method is based upon a **complex-frequency analysis** of numerical dispersion relation (21).
In the previous section, we only use a complex-wavevector analysis (using $\tilde{k}$), below both it is $\omega$ that becomes a complex quantity.


#### II.4.2 - Complex-frequency analysis
We consider the complex exponential traveling waves like in Eq. (3) present in our finite-difference grid and discretely sampled at ($x_i, t-n$). But now, we allow for the possibility of a complex-valued numerical angular frequency such that $\tilde{\omega} = \tilde{\omega_{real}} + i \tilde{\omega_{imag}}$. Eq. (18) gives:


$u^n_i = e^{i(\tilde{\omega} t_n - \tilde{k} x_i)}$


$u^n_i = e^{i(\tilde{\omega} n \Delta t - \tilde{k} i \Delta x)} = e^{-\tilde{\omega}_{imag} n \Delta t}e^{i(\tilde{\omega}_{real} n \Delta t - \tilde{k} i \Delta x)}$


Following a similiar procedure like the one followed for the numerical dispersion study, we get:


$cos(\tilde{\omega} \Delta t) = (\frac{c \Delta t}{\Delta x})^2 (cos(\tilde{k} \Delta x)-1)+1$


Inverting it and introducing the Courant stabiity factor $S= \frac{c \Delta t}{\Delta x}$, one gets:


$\tilde{\omega} = \frac{1}{\Delta t} arccos(S^2(cos(\tilde{k} \Delta x)-1)+1)$


$\tilde{\omega} = \frac{1}{\Delta t} arccos(\xi) = \frac{1}{\Delta t} (\frac{\pi}{2} - arcsin(\xi))$
with $\xi = S^2(cos(\tilde{k} \Delta x)-1)+1$

From Eq. (49), we obtain: $1 - 2S^2 \leq \xi \leq 1$ for all $\tilde{k}$ $\epsilon$ $\Re$.

Now, we consider divinding the $\xi$ range defined by the previous inequality in:

>1) $-1 \leq \xi \leq 1$ This subrange exists for $0 \leq S \leq 1$. Here, $arcsin(\xi)$ is real-valued, and hence, real values of $\tilde{\omega}$ are obtained in Eq. (48). With $\tilde{\omega}_{image}=0$, Eq. (45) yields a **constant wave amplitude** with time. This is the **case of pratical interest for useful numerical simulations.** 


>2) $1 - 2S^2 \leq \xi \leq -1$ This subrange exists for $1 - 2S^2 \leq -1$, or equivalently for $S>1$. The most negative value in this subrange, $\xi_{lowerbound}$ occurs for $cos(\tilde{k} \Delta x) = -1$, i.e. $\tilde{k} \Delta x = \pi$. We have $\xi_{lowerbound} = 1 - S^2$ for $\tilde{k} \Delta x = \pi$.

### II.5 - The [Yee's algorithm](chrome-extension://efaidnbmnnnibpcajpcglclefindmkaj/https://empossible.net/wp-content/uploads/2020/08/Lecture-Introduction-to-the-Yee-Grid.pdf)
#### II.5.1 - Introduction

The basic idea of the seminal Yee's algorithm, published in 1966, is to locate electromagnetic field components in **specified positions in a unit 3D cell** in order to generate a finite-difference scheme of Maxwell's equations capable of calculating its **time and space derivatives with second-order accuraby**.



#### II.5.2 - Description
1) The Yee's algorithm **solves for both electric and magnetic fields in time and space** using the coupled Maxwell's curl equations, rather than solving for the electric field alone (or the magnetic field alone) with a wave equation.

>- 1. This is analogous to the combined-field integral equation formulation of the moment methods, wherein both **E** and **H** boundary conditions are enforced on the surface of a material strucure.
>- 2. Using both **E** and **H** information, the solutions is more robust than using either alone (i.e, it is accurate for a wider class of structures). Both electric and magnetic material propoerties can be modeled in a straightforward manner. 
>- 3. Features uniques to each field, such as tangential **H** singularities near edges and corners, azimuthal (looping) **H** singularities near thin wires, and radial **E** singularities near points, edges and thin wires, can be individually modeled if both electric and magnetic fields are available.

2) As illustrated in Fig. (7) and (8), the Yee's algorithm centers its **E**
 and ***H*** components in a three-dimensional space so that every **E** component is surrounded by four circulating **H** components, and every **H** components is surrounded by four circulating **E** components.

>> This provides a simple picture of three-dimensional space being filled by an **interlinked array** of Faraday's law and Ampere's law contours.

>> For example, it is possible to identity Yee **E** components associated with displacement current flux linking **H** loops, as well as **H** components associated with magnetic flux linking **E** loops.


>> Actually, the Yee algorithm simultaneously simulates the pointwise differential form and the macroscopic integral form of Maxwell's equations. The latter is extremely useful in specifying field boundary conditions and singularities. In addition, we have the three following attributes of the Yee's space lattice:



>- 1. The finite-difference expressions for the space derivatives used in the curl operators are central-difference in nature and second-order accurate.
>- 2. Continuity of tangential **E** and **H** is naturally maintained across an interface of dissimilar materials if the interface is parallel to one of the lattice coordinate axes. For this case, there is no need to specially enforce field boundary conditions at the interface. At the beginneing of the problem, we simply specify the material permittivity and permeability at each field component location. This yield a stepped or "staircase" approximation of the surface and internal goemetry of the structure, with a space resolution set by the size of the lattice unit cell.
>- 3. The loction of the **E** and **H** components in the Yee's space lattice and the central-difference operations on these components implicitly **enfore the wo Gauss' law relations**. This, the **Yee's mesh is divergence-free** with respect to its **E** and **H** fields inthe absence of free electric and magnetic charge.

3) As illustrated in Fig. (9) , the Yee's algorithm also centers its **E** and **H** components **in time**, in what is termed a **leapfrog arrangement**.

>- All of the **E** computations in the modeled space are completed and stored in memory for a particular time point using previously stored **H** data.
>- Then, all of the **H** computations in the space are completed and stored in memory using the **E** data just computed.
>- The cycle begins agains with the recomputation of the **E** components based on the newly obtained **H**.
>- This process continues until time-stepping is concluded.

> 1. Leapfrog time-stepping is **fully explicit**, thereby avoiding problems involved with simultaneous equations and matrix inversion.

> 2. **The finite-difference expressions for the time derivatives are central-difference in nature and second-order accurate.**

> 3. **The time-stepping algorithm is nondissipative for correct values of space and time discretization steps**. That is, numerical wave modes propagating in the mesh do not spuriously decay due to a nonphysical artifact of the time-stepping algorithm.

$\mathbf{D}(\mathbf{r},t) = \epsilon(\mathbf{r})\mathbf{E}(\mathbf{r},t)$

### III.3 - Complex coordinate stretching
We consider $\mathbf{w}(\mathbf{x},t)$ the solution of some wave equations in a *unbounded domain*.
We want to truncate the space outside the region of interest (ROI) in such a way to absorb radiating waves.

This truncation is realized in 3 main steps:

1 - *Analytic continuation* of the solutions and equations to a *complex-x*-contour which transforms oscillating waves into **exponentially decreasing waves** outside the region of interest *without reflections*.

2 - *Coordinate transformation* to express the complex $x$ as a function of real coordinate.
We have moved from *real coordinates* to *complex material regions.*

3 - *Truncation* of the new real coordinate system inside the complex-material region. The solution being decreasing in this last region, we can truncate it after a sufficiently long distance.

We assume:

- the space far from the ROI is homogenous.
- the space far from the ROI is linear and time invariant.


Consequently, the *radiating solutions* in infinite space must take the form of a superposition of plane waves:


$\mathbf{W}(\mathbf{x},t) = \int_{\mathbf{k},\omega} \mathbf{W}_{\mathbf{k},\omega}e^{i(\mathbf{k}\cdot \mathbf{x}-wt)}$


The key property is that the *radiating solutions* (along + x-direction) can be decomposedinto functions of the form:


$\mathbf{W}(y,z)e^{i(k_x\cdot x-wt)}$

### III.4 - Consequences of the complex coordinate stretching


- The solution is not changed before the complex coordinate streching (x<0).

### III.7 - Coordinate transformations and modified materials
- It can be shown that the $1+i\sigma_x(x)/\omega$ term introduced by the PML coordinate transformation is equivalent to introduce an *artificial anisotropic absorbing material* in the region surrounding the ROI of the computational domain instead of the initial material.
- In eq. (63), $a$ and $b$ become complex and $a$ becomes a tensor to tackle the anistropy along the $x$,$y$ and $z$-directions.
- This behavious is an example of a more general result:
**Maxwell's equations under any coordinate transformation can be expressed as Maxwell's equations in Cartesian coordinates with** *transformed materials*.


This is known under the name *transformation optic principle*


More explicitely:
Details of the calculues to get $\Sigma = diag(\frac{S_yS_z}{S_x},\frac{S_xS_z}{S_y},\frac{S_xS_y}{S_z})$

-  $\Sigma = \mathbf{J}\mathbf{J}^2/det(\mathbf{J})$ and $\mathbf{J} = diag(S^{-1}_x,S^{-1}_y,S^{-1}_z)$ so
-  $\Sigma = diag(S^{-2}_x,S^{-2}_y,S^{-2}_z)/diag(S^{-1}_x,S^{-1}_y,S^{-1}_z)$ so
-  $\Sigma = diag(\frac{S_yS_z}{S_x},\frac{S_xS_z}{S_y},\frac{S_xS_y}{S_z})$

- Let us consider a map form the coordinate system {x,y,z} to the coordinate system {x',y',z'} given by the function x{x',y',z'}, y{x',y',z'} and z{x',y',z'}.
- All the information is in the transformation of the differentials and is therefore given by the chain rule:


$dx = \frac{\partial{x}}{\partial{x'}}dx'+\frac{\partial{x}}{\partial{y'}}dy'+\frac{\partial{x}}{\partial{z'}}dz'$


$dy = \frac{\partial{y}}{\partial{x'}}dx'+\frac{\partial{y}}{\partial{y'}}dy'+\frac{\partial{y}}{\partial{z'}}dz'$


$dz = \frac{\partial{z}}{\partial{x'}}dx'+\frac{\partial{z}}{\partial{y'}}dy'+\frac{\partial{z}}{\partial{z'}}dz'$


All the information is in the **Jacobian matrix J**

$dxdydz = \mathbf{J}dx'dy'dz'$


$J(x',y',z') = \frac{\partial{(x,y,z)}}{\partial{(x',y',z')}} = \begin{bmatrix}
\frac{\partial{x}}{\partial{z'}} & \frac{\partial{x}}{\partial{y'}} & \frac{\partial{x}}{\partial{z'}} \cr
\frac{\partial{y}}{\partial{x'}} & \frac{\partial{y}}{\partial{y'}} & \frac{\partial{y}}{\partial{z'}} \cr
\frac{\partial{z}}{\partial{x'}} & \frac{\partial{z}}{\partial{y'}} & \frac{\partial{z}}{\partial{z'}}
\end{bmatrix}$


It can be shown that scalar quantitites like the permittivities and the permittivities and permeabilites transform themselves as:

$\underline{\underline{\epsilon'}}(\mathbf{x}) = \mathbf{J}(\mathbf{x'})\underline{\underline{\epsilon}}(\mathbf{x}(\mathbf{x}'))\mathbf{J}^T(\mathbf{x}')/det(J(\mathbf{x}'))$


$\underline{\underline{\mu'}}(\mathbf{x}) = \mathbf{J}(\mathbf{x'})\underline{\underline{\mu}}(\mathbf{x}(\mathbf{x}'))\mathbf{J}^T(\mathbf{x}')/det(J(\mathbf{x}'))$

where the T exponant means transposed.
In our simple case for the PML in Cartesian coordinates, we have:


$\mathbf{J} = diag(S^{-1}_x,S^{-1}_y,S^{-1}_z)$ with $S_u=1+i\sigma_u/\omega$ for $u = x,y,z$


All isotropic tensor $\underline{\underline{\xi}}$ commutes with diagonal $J$ as a result we can simplify eq. (79), therefore for isotropic $\underline{\underline{\epsilon}}$ (focusing on the permittivity only), one gets:


$\underline{\underline{\epsilon'}} = \Sigma \underline{\underline{\epsilon}}$ with $\Sigma \equiv \mathbf{JJ}^T/det(\mathbf{J})=diag(\frac{S_yS_z}{S_x},\frac{S_xS_z}{S_y},\frac{S_xS_y}{S_z})$

### III.8 - PML in the time domain
One question arises: How to implement PML in the time domain since the term $1/\omega$ can not be computed directly?


The answer is to use the *auxiliary differential equation* (ADE) approach.


Let us consider the scalar wave equation 1D assuming the harmonic regime time dependency in $e^{i \omega t}$.


$\frac{\partial{u}}{\partial{t}} = b \frac{\partial{\nu}}{\partial{x}} = -i \omega u$


$\frac{\partial{\nu}}{\partial{t}} = a \frac{\partial{u}}{\partial{x}} = -i \omega \nu$


Then, we perform, the PML transformation corresponding to eq. (74):


$b \frac{\partial{\nu}}{\partial{x}} = -i \omega u + \sigma_x u$


$a \frac{\partial{u}}{\partial{x}} = -i \omega \nu + \sigma_x \nu$

We note that the $1/\omega$ term has vanished. We now apply the inverse Fourier transform to get time-domain expressions:

$\frac{\partial{u}}{\partial{t}} = b \frac{\partial{\nu}}{\partial{x}} -\sigma_x u$


$\frac{\partial{\nu}}{\partial{t}} = a \frac{\partial{u}}{\partial{x}} -\sigma_x \nu$


With $\sigma_x > 0$:
- if both $u$ and $\nu$ are constant in space, then the solutions decay in time
- if both $u$ and $\nu$ are constant in time , then the solutions decay in space
>- $\sigma_x$ plays the role of a conductivity (electric and magnetic one)


The scalar wave equation written in 2D and split using the auxiliary field $\mathbf{\nu}=[\nu_x \nu_y]$ is :


$\frac{\partial{u}}{\partial{t}} = b\nabla \cdot \mathbf{\nu} = -b\frac{\partial{\nu_x}}{\partial{x}} + \frac{\partial{\nu_y}}{\partial{y}}=i\omega u$



$\frac{\partial{\nu_x}}{\partial{t}} = a\frac{\partial{u}}{\partial{x}} = -i \omega \nu_x$



$\frac{\partial{\nu_y}}{\partial{t}} = a\frac{\partial{u}}{\partial{y}} -i \omega \nu_y$


Using the PML transformation (74), in the two first equations, we obtain:


$\frac{\partial{u}}{\partial{t}}        = \frac{b}{1+i\sigma/\omega}\frac{\partial{\nu}}{\partial{x}}+b\frac{\partial{\nu}}{\partial{y}}$


$\frac{\partial{\nu_x}}{\partial{t}}    = \frac{a}{1+i\sigma/\omega}\frac{\partial{u}}{\partial{x}}$


$\frac{\partial{\nu_y}}{\partial{t}}    = a\frac{\partial{u}}{\partial{y}}$


Then multiplying by (1+i\sigma/\omega) and computing explicitely the time derievative using the harmonic regime hypothesis, we get for the forst 2 equations:


$b\frac{\partial{\nu_x}}{\partial{y}} + b\frac{\partial{\nu_y}}{\partial{y}} (1+i\sigma/\omega)= -i \omega u + \sigma_x u$


$a\frac{\partial{u}}{\partial{x}}=-i \omega \nu + \sigma_x \nu$
- We can inverse-Fourier transform to the time domain the above equation $-i\omega \Rightarrow{} \frac{\partial{}}{\partial{t}}$.
- But the term $b\frac{\partial{\nu_y}}{\partial{y}}(i \sigma_x/\omega)$ contains a $1/\omega$ factor. $-i/\omega \Rightarrow{} \int ... \partial{t}$


We need to introduce a new auxiliary scalar field $\psi$ such that:


$i \omega \psi \equiv b \sigma_x \frac{\partial{\nu_y}}{\partial{y}}$ with $\psi(0)=0$


So we obtain:


$b\frac{\partial{\nu_x}}{\partial{y}} + b\frac{\partial{\nu_y}}{\partial{y}}  - \psi = -i \omega u + \sigma_x u$


We again inverse Fourier transform everything in the time domain. We obtain a set of 4 time-domain equations with PML in the $x$-direction:


$\frac{\partial{u}}{\partial{t}}        =b\nabla \cdot \nu - \sigma_x u - \psi$


$\frac{\partial{\nu_x}}{\partial{t}}    =a \frac{\partial{u}}{\partial{x}}- \sigma_x \nu_x$


$\frac{\partial{\nu_y}}{\partial{t}}    =a \frac{\partial{u}}{\partial{y}}$


$\frac{\partial{\psi}}{\partial{t}}     =b \sigma_x\frac{\partial{\nu_y}}{\partial{y}}$


We can notice that the PML is equivalent to consider an *anisotropic absorber* where the real scalar $a$ is replaced by a 2x2 complex tensor:


$\begin{bmatrix}
\frac{1}{1+i\sigma/\omega} & 0 \cr
0 & 1
\end{bmatrix}$


This is an example of general property linking coordinate transformations and materials given in Sec. 133.


To demonstrate the previous result, we plug the transformed  expression of $a$ eq. (83) without evaluating the time derivative:


$\frac{\partial{u}}{\partial{t}}        =\frac{b}{1+i\sigma/\omega}\frac{\partial{\nu_x}}{\partial{x}}+\frac{\partial{\nu_y}}{\partial{y}}$


$\frac{\partial{\nu_x}}{\partial{t}}    =\frac{a}{1+i\sigma/\omega}\frac{\partial{u}}{\partial{x}}$


$\frac{\partial{\nu_y}}{\partial{t}}    =a\frac{\partial{u}}{\partial{y}}$


This system is exactly the one obtained after the PML transformation described by eq. (74).
*We note that b must also be transformed such taht the initial term* $b\nabla$ must be replace by a term $\mathbf{b} \cdot \nabla$ where $\mathbf{b}$ is a vector $[\frac{b}{1+i\sigma/\omega}b]$

### III.9 - PML for Maxwell's equation in the harmonic regime
#### III.9.1 - Coordinate strectching
Maxwell's equation in the international unit system are (in this part about PML we use the $\nabla$ notation instead of the $\mathbf{curl}$ and $\mathbf{grad}$ used in the FDTD and FEM parts):


$\nabla \cdot \mathbf{D}(\mathbf{r},t) = \rho $


$\nabla \cdot \mathbf{B}(\mathbf{r},t) = 0 $


$\nabla \times \mathbf{E}(\mathbf{r},t) = -\frac{\partial{\mathbf{B}}(\mathbf{r},t)}{\partial{t}}$


$\nabla \times \mathbf{H}(\mathbf{r},t) = \frac{\partial{\mathbf{D}}(\mathbf{r},t)}{\partial{t}} + \mathbf{J}$


Without any charge and current (source-free condition) they give:


$\nabla \cdot \mathbf{D}(\mathbf{r},t) = 0 $


$\nabla \cdot \mathbf{B}(\mathbf{r},t) = 0 $


$\nabla \times \mathbf{E}(\mathbf{r},t) = -\frac{\partial{\mathbf{B}}(\mathbf{r},t)}{\partial{t}} $


$\nabla \times \mathbf{H}(\mathbf{r},t) = \frac{\partial{\mathbf{D}}(\mathbf{r},t)}{\partial{t}} $


In the harmonic regime with a $e^{i \omega t}$ time dependency, one obtains after removing all time-dependent exponential:


$\nabla \cdot \mathbf{D}(\mathbf{r}) = 0 $


$\nabla \cdot \mathbf{B}(\mathbf{r}) = 0 $


$\nabla \times \mathbf{E}(\mathbf{r}) = i \omega\mathbf{B}(\mathbf{r}) $


$\nabla \times \mathbf{H}(\mathbf{r}) = -i \omega\mathbf{D}(\mathbf{r}) $


using the simple local constitutive relations $\mathbf{D}(\mathbf{r}) = \epsilon \mathbf{E}(\mathbf{r})$ and $\mathbf{B}(\mathbf{r})=\mu \mathbf{H}(\mathbf{r})$ ,one gets:

$\nabla \cdot \epsilon \mathbf{E}(\mathbf{r}) = 0 $


$\nabla \cdot \mu \mathbf{H}(\mathbf{r}) = 0 $


$\nabla \times \mathbf{E}(\mathbf{r}) = i \omega \mu \mathbf{H}(\mathbf{r}) $


$\nabla \times \mathbf{H}(\mathbf{r}) = -i \omega \epsilon \mathbf{E}(\mathbf{r}) $


We consider the modified Maxwell's equations through the use of the change of the spatial derivative as given in eq. (74) to implement the complex coordinate stretching:


$\frac{\partial{}}{\partial{x}} \Rightarrow{}\frac{1}{1+i\sigma_x(x)/\omega}\frac{\partial{}}{\partial{x}}= \frac{1}{S_x(x)}\frac{\partial{}}{\partial{x}} $


We define a modified $\nabla$ operator denoted by $\nabla_s$ such that:


$\nabla_s = \frac{1}{S_x(x)}\frac{\partial{}}{\partial{x}}\mathbf{e_x} + \frac{1}{S_y(y)}\frac{\partial{}}{\partial{y}}\mathbf{e_y} + \frac{1}{S_z(z)}\frac{\partial{}}{\partial{z}}\mathbf{e_z}$


Using $\nabla_s$, we can rewrite the Maxwell's eqs including the complex coordinate stretching:


$\nabla_s \cdot \epsilon \mathbf{E}(\mathbf{r}) = 0 $


$\nabla_s \cdot \mu \mathbf{H}(\mathbf{r}) = 0 $


$\nabla_s \times \mathbf{E}(\mathbf{r}) = i \omega \mu \mathbf{H}(\mathbf{r}) $


$\nabla_s \times \mathbf{H}(\mathbf{r}) = -i \omega \epsilon \mathbf{E}(\mathbf{r}) $


To study what occurs at the interface between the original medium and the stretched one, let us consider a plane wave such that:


$\mathbf{E} = \mathbf{E}_0 e^{i\mathbf{k}\cdot \mathbf(r)} = \mathbf{E}_0 e^{i (k_x\cdot x + k_y\cdot y + k_z\cdot z)}$


$\mathbf{B} = \mathbf{B}_0 e^{i\mathbf{k}\cdot \mathbf(r)} = \mathbf{B}_0 e^{i (k_x\cdot x + k_y\cdot y + k_z\cdot z)}$


We substitute these two fields ine the modified Maxwell's equation (118) to get:


$\mathbf{k}_s \cdot \mathbf{E} = 0$


$\mathbf{k}_s \cdot \mathbf{H} = 0$


$\mathbf{k}_s \times \mathbf{E} = \omega \mu \mathbf{H}$


$\mathbf{k}_s \times \mathbf{H} = -\omega \epsilon \mathbf{E}$


with $\mathbf{k}_s = \frac{k_x}{S_x}\mathbf{e_x} + \frac{k_y}{S_y}\mathbf{e_y} + \frac{k_z}{S_z}\mathbf{e_z}$


From the third equation we compute:


$\mathbf{k}_s \times (\mathbf{k}_s \times \mathbf{E})= \omega \mu (\mathbf{k}_s \times \mathbf{H}) = -\omega^2 \mu \epsilon \mathbf{E}$


Using vector formula (277) in Appendix, we obtain:


$(\mathbf{k}_s \cdot \mathbf{k}_s)\mathbf{E} = \omega^2 \mu \epsilon \mathbf{E}$


So we finally get the dispersion equation:


$mathbf{k}_s \cdot \mathbf{k}_s = \omega^2 \mu \epsilon = \kappa^2$ or equivalently


$ (\frac{k_x}{S_x})^2+(\frac{k_y}{S_y})^2+(\frac{k_z}{S_z})^2= \kappa^2$


The solution to this dispersion is:


$k_x = \kappa S_x cos(\theta)$


$k_y = \kappa S_y sin(\theta)cos(\phi)$


$k_z = \kappa S_z sin(\theta)sin(\phi)$


Let us compute the *impedance* $\eta$:


$\eta= \frac{|\mathbf{E}|}{|\mathbf{H}|} = \frac{\omega \mu}{|\mathbf{k_s}|} = \frac{|\mathbf{k_s}|}{\omega \epsilon} = \sqrt{\frac{\mu}{\epsilon}}$


We see that **the stretching process does not change the wave impedance** consequently no part of the incident plane wave will be reflected at the interface between the original material and the on with the complex coordinate stretching.

### IV.1 - Basics of Finite Element Method
#### IV.1.1 - Introduction

The finite element method can be seen as a general discretization procedure of continuum problems posed by mathematically defined statements.

- 1. The finite element method (FEM) is extremely general as it allow the treatment of large class of *partial differential problems*, including non-linear problems and without any limitation on the geometry involved.
- 2. This method is based on an *integral formulation* derived from the partial differential equations coupled with the boundary conditions. Several integral formulations are possible but in this lecture we will deal with only the *projection ones*, and more precisely the *weighted residual one*.
- 3. The obtaining of the exact function of the integral formulation is as difficult as the partial differential problems. To reduce this difficulty, it is an *approximated solution expressed as a linear combination of known independent functions* that will be looked for.
Usually, these functions are *priecewise polynomial functions* in the FEM described in this lecture.
- 4. In the FEM, the domain of interest will be split in sub-domains called *elements* such that in each of them the unknown function is approached by a polynom.
- 5. Finally, the FEM replaces the system of partial differential equations with a *system of equations for the coefficients of the approximated function* 
- 6. In the last thirty years, the FEM in electromagnetism has not only been improved but also better understood and, nowaday, the reasons why some approaches work much better than others are quite clear.
- 7. In the case of propagation mode computation, the strategy of the finite element method is quite straightforward: a matrix (i.e. a finite rank operator) that approximates the partial  differential operator of the problem is constructed numerically and a numerical eigenvalue problem is solved using numerical algebra techniques.

#### IV.1.4 - A 1D naive introduction

The general principles are quite simple and the main ingredients are presented below in the context of detailed one-dimensional toy problem:
- 1. the PDE boundary value problem is defined.
- 2. the unknown function $u(x)$ is approximated by linear combinations of shape functions.
- 3. Weighted residuals are introduced within a integral formulation to compute the coefficient of the linear combination.
- 4. Weak form and boundary conditions are defined and used.
- 5. the linear system to be solved is written.


##### IV.1.4.1 - The **boundary value problem** to solve is a differential equation:


$L u(x) = \frac{d}{dx}(p(x)\frac{du(x)}{dx})+q(x)u(x)=f(x)$

where $p(x)$, $q(x)$ and $f(x)$ are given functions and $u(x)$ is the **unknown function**. The geometrical domain is merely and interval [$a,b$] and the boundary conditions are taken here to be *homogeneous Dirichlet conditions* $u(a)=0$ and $u(b)=0$ which ensure the uniqueness of the solution.
The operator L is *self-adjoint* and it is a problem of the *Sturm-Liouville* type (see Appendix 1).


Remark: This is very different to a Cauchy initial value problem, in which two conditions for this 2nd order ODE are given on a, i.e. the value of $u$ and its derivative. Both types of conditions provide 2 constraints as needed to solve the problem.


##### IV.1.4.2 - The unknown function $u(x)$ is approximated by **linear combination**


$\tilde{u}(x) = \Sigma u_i \phi_i (x)$


of simple function $\phi_i(x)$ called *shape functions* that form a basis of a finite dimensional functional space.


Remark: There are several ways to choose such functions. Choosing for instance, sine and cosine functions leads to the Fourier method in the approximation is in this case a truncated Fourier series. This is an extremely useful method but this approch has two drawbacks. The first one is the the approximation is smooth ($\epsilon C^{\inf} [a,b]$) and a large number of terms is necessary if there are discountinuities in the $p(x)$, $q(x)$ and $f(x)$ functions, and Gibb's phenomenon can not fully avoided.


The second drawback is that the generalisation to several dimensions is easy only with simple shapes for the geometrical domain, e.g. a rectrangle or a disk in two dimensions. In practice, if you decide to adopt this approach, you will get in the end a variant of the *plane wave method*. This method is no longer described in this lecture, it was described in the old lecture notes on computational photonics, see Joannopoulos's [book](https://pdfs.semanticscholar.org/d79e/59a67b1e7e7b335f8b8b5d813bd46722bb23.pdf) for some properties.


A common choice to use for the approximation of the searched function is a piecewise linear function i.e. a linear combination of "hat" functions. Given the nodes $x_i$, the end points of the subintervals such tat $x_0 = a < x_1 < ... < x_i  < ...  < x_n=b$ the hat function associated with $x_i$ is defined by:


$\phi_i(x) = \frac{x-x_{n+1}}{x_i-x_{i+1}}$ $x \epsilon [x_i,x_{i+1}]$


$\phi_i(x)=\frac{x-x_{n-1}}{x_i-x_{i-1}}$ $x \epsilon [x_{i-1},x_i]$ for $i = 1, ... n-1$


$\phi_i(x)=0$ elsewhere,


$\phi_0(x)=\frac{x-x_1}{x_0-x_1}$ if $x \epsilon [x_0,x_1]$ and elsewhere,


$\phi_n(x)=\frac{x-x_{n-1}}{x_n-x_{n-1}}$ if $x \epsilon [x_{n-1},x_n]$ and elsewhere,


With this choice, the coefficients $u_i$ in Eq. (156) are the *nodal values* of the approximations.

3. In order to determine the coefficient $u_i$, weighted residuals:


$R(u,w) \equiv \int^b_a(Lu(x)-f(x))w(x)dx$


are built by integrating the product of the equation to be solved $Lu-f=0$ with a *weight function* w(x), integrating of course on the domain of definition of the problem $[a,b]$.


For the solution $u$, the weighted residuals are equal to zero for any weight function since the integrand is equal to zero: $R(u,w)=0$. A numerical solution is obtained by introducing the *approximation function* (see Formula (156)) $\tilde{u}(x)=\Sigma^{n-1}_{i=1}u_i\phi_i(x)$ in the weighted residuals: 


$R(\tilde{u}, w) = \Sigma u_i \int^b_a(L\phi_i(x))w(x)dx - \int^b_af(x)w(x)dx = 0$ 


and choosing a suitable set of weight function. **The most common choice for the weight functions is to take the shape functions themselves**. This leads to the *Galerkin method* where the equations to solve are:


$R(\Sigma u_i \phi_i, \phi_j) = \Sigma u_i \int^b_a(L\phi_i(x))\phi_j(x)dx - \int^b_af(x)\phi_j(x)dx = 0$ 


Remark:
Other methods exist within the finite element approach are not based on weighted residuals. For example the Ritz method is based on a functional whose minimum corresponds to the partial differential problem under investigation. In simple cases, the Ritz method and the Galerkin method of the weighted residual approach provide the same system to be solved

4. This is extremely simple but there is a pitfall: the evaluation of the residue involves a second derivative with respect to $x$ which makes no sense for the piecewise linear approximiation since it is equal to zero (the first derivative is piecewise constant and discontinuous).


**Rather than taking distributional derivatives, the common solution is to introduce the weak formulation obtained via an integration by parts**:


$\int^b_a \frac{d}{dx}(p(x)\frac{du(x)}{dx}w(x)dx) = - \int^b_a{p(x)\frac{du(x)}{dx}\frac{dw(x)}{dx}dx} + [p(x)\frac{du(x)}{dx}w(x)]^b_a$


The last term that appears involves the boundary conditions. In the present example, homogeneous Dirichlet conditions have to be imposed (u(a)=u(b)=0). This is easily done by taking an approximiation satisfying these *essential conditions*: the nodal values corresponding to $x=a$ and  $x=b$ are equal to zera, the weight functions are the shape functions of the internal nodes (i.e. excluding the two boundary nodes) only so that  $w(a)=w(b)=0$ and the boundary ter cancels.


Remark: If ones keeps the nodal values on the boundary as unknowns so that the approximation is now $\tilde{u}(x) = \Sigma^n_{i=0}u_i\phi_i(x)$(take note of the limits of the summation) and introduces the corresponding shape functions $\phi_0$ and $\phi_n$ in the set of weight functions $w$ then the problem corresponds to a problem with homogeneous *Neumann conditions or natural conditions* $\frac{du}{dx}|_{w=a=0}$ and $\frac{du}{dx}|_{w=b=0}$ on the boundary. Once again, two conditions are needed.


The residual can be expressed as:


$\int^b_a{(p(x)\frac{du(x)}{dx}\frac{dw(x)}{dx} + q(x)u(x)w(x))dx} - \int^b_a f(x)w(x)dx = 0$


Introducing the **symmetric bilinear form**:


$A(u,w) \equiv \int^b_a{(p(x)\frac{du(x)}{dx}\frac{dw(x)}{dx} + q(x)u(x)w(x))dx}$


and writing


$\left<f|w \right> \equiv \int^b_a f(x)w(x)dx$


to define a duality product, Eq (163) can be expressed in a very concise abstract form:


$A(u,\mathbf{w}) = \left<f|w \right>$


It appears that the constraints on the differentiability of $u$ are now weaker, satisfied by piecewise linear functions, and have been partially transferred on $w$. The two functions play a symmetrical role which suggest that the Galerkin approach is quite natural. In terms of functional spaces, the weak solution is in $H^1([a,b])$, though the differential problem involves a second order operator and is defined by:


$A(u,\mathbf{w}) = \left<f|w \right>$, ∀$\mathbf{w}$ $\epsilon$ $H^1([a,b])$


(See Sobolev and Hilbert spaces Sections 4 page 289 in appendix). The numerical solution is in the finite diemnsional subspace $V_h C H^1([a,b])$ spanned by the $\phi_i$ The approximation $\tilde{u}$ given in Eq. (156) is defined by:


$A(\tilde{u},\phi) = \left<f|\phi \right>, ∀$\phi$ $\epsilon$ $V_h$


5. The equality above is satisfied if and only if it holds for all the basis functions $\phi_j$ and this gives the following linear  system of equations:


$\Sigma u_i A(\phi_i,\phi_j) = \left<f|\phi_j \right>$,for $j=1,...,n-1$


This linear system is sparse since only two nodes which are the end points of the same subinterval give a nonzero coefficient in the system matrix.


We can rewrite the previous equation in a matrix form, using a column vector notation for the $u_i$:


$[\mathbf{A}][u]=[\mathbf{F}]$


The matrix $\mathbf{A}$ is often called $\mathbf{K}$ in many textbooks and articles, we will use this notation in the following sections:


#### IV.1.6 - Numerical dispersion and accuracy versus element order

Let us study a plane linear wave propagating in a homogeneous bulk medium along the $x$-axis, due to the numerical space discretization, the wave simulated with the FEM will propagate with a wavenumber $\tilde{k}$ different from the theoretical one denoted by $k$.


This difference is called the *numerical dispersion*.

Below we evaluate this dispersion for a 1D wave with nodes equally spaced by a distance *h* in the FEM.

The wave equation to solved is the scamar Helmotz equation 


$\frac{d^2\phi(x)}{dx^2} + k^2\phi(x)=0$


This equations is a special case of Eq. (155) seen at the beginning of the lecture:


$p(x) = 1$, $q(x) = k^2$ and $f(x) = 0$


Taking $p(x) = q(x) = 1$ and $f(x) = 0$ Consequently, we can apply and simplify all the formula we have seen before.
The theoretical wave can be expressed as:


$\phi(x) = A e^{ikx}$ where A is the amplitude of the wave. The FEM solution of this wave have the form in the case of equality spaced elements by a length $h$:


$\phi_i = \phi(x_i) = \phi_a e^{i \tilde{k}x_i} = \phi_a e^{i \tilde{k}ih}$


where $\phi_a$ is the numerical amplitude of the wave at node 0.
To evaluate $\tilde{k}$ we need tp evaluate the elements of the $\mathbf{K}$ (see Eq. 171). Using the definition of symmetric bilinear from $a$ given by Eq. (167), simplifying it equation using Eq. (181), one gets: 


$\int^b_a(\frac{d\phi(x)}{dx}\frac{dw(x)}{dx} - k^2 \phi(x)w(x))dx$


Then using the approximation for $\phi$ using the shape function $N_i$, and the frame of the Galerkin approach through Eq. (169), one obtains for each element $K_{ij}$ of the matrix $\mathbf{K}$:


$K_{ij} = \int^b_a(\frac{dN_i(x)}{dx}\frac{dN_j(x)}{dx} - k^2 N_i(x)N_j(x))dx$


This integral can be computed analytically and easily when the shape function $N_i$ are the linear hat shape functions given by Eqs. (157). We gets assumng  equally space nodes at distance $h$:


$K_{ij} = (1/h+1/h)-\tilde{k}^2(h/3+h/3)$


$K_{i+1,j} = K_{i,j+1} = -(1/h)-\tilde{k}^2(h/6)$


Now, we can rewrite the system (171):


$K_{i,j-1}\phi_{i-1} + K_{i,j}\phi_{i} + K_{i,j+1}\phi_{i+1}=0$


evaluating the $K_{ij}$ coefficients in it:


$-(1/h+k^2h/6)\phi_i + 2(1/h-k^2h/6)\phi_{+1}=0$


Then injecting Eq. (183) to get:


$(1/h-k^2/6)cos(\tilde{k})=(1/h+k^2h/3)$


We finally have the analytical expression of the numerical wavenumber in our 1D scalar wave equation problem:


$\tilde{k} = \frac{1}{h} arccos(\frac{6-2(kh)^2}{6-(kh)^2})$


We can approximate this relation keeping only the first two terms in the Taylor expansion of the cosine function to compute the relative error of the numerical wavenumber computed from the FEM with linear shape functions:


$\frac{\tilde{k}-k}{k} \approx \frac{1}{12}(kh)^2 = \frac{\pi^2}{3}(\frac{h}{\lambda})^2$


where $k=2\pi/\lambda$

-This equation show that the numerical dispersion decreases quadratically with the size of the elements for linear shape functions $(\frac{h}{\lambda})^2$.
- It can be shown that if the quadratic element are used (see Eqs. (173)) then the numerical dispersion decreases at $(\frac{h}{\lambda})^4$ meaning that it is more efficient to increase the order of the shape functions to reduce the numerical dispersion that to reduce the size of the finite elements.
- If $p$-order shape functions are used, then the numerical dispersion scales as $(\frac{h}{\lambda})^{2p}$. This is one of the key advantages of the FEM compared to usual finite difference method.

$ \mathbf{v}  = v_x(x,y) e^x + v_y(x,y) e^y + v_z(x,y) e^z = \mathbf{v}_t + v_z e^z$ with transverse part given by the vector field $\mathbf{v}_t = v_x(x,y) e^x + v_y(x,y) e^y$ and the longitudianl (axial) component given by the scalar function $ v_z(x,y)$

#### IV.2.2 - Vector operators


We derine the transverse operators:


$\mathbf{grad}_t v_z = \frac{\partial{v_z}}{\partial{x}}e^x + \frac{\partial{v_z}}{\partial{y}}e^y$


$\mathbf{curl}_t v_t = \frac{\partial{v_y}}{\partial{x}} - \frac{\partial{v_x}}{\partial{y}}$


$\mathbf{div}_t v_t = \frac{\partial{v_x}}{\partial{x}} + \frac{\partial{v_y}}{\partial{y}}$

$\mathbf{curl}_{\beta} \mathbf{v} = \mathbf{curl}_t v_t + (\mathbf{grad}_t v_z - i \beta \mathbf{v}_t) \times e^z$


It is important here to bear in mind that we are in the harmonic case ($e^{-i \omega t}$ time dependency), dealing with complex valued fields, and that *the constant* $\beta$ may be *also complex valued* (leaky modes).
